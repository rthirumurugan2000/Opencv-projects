### Webcam implementation

```python
cap = cv2.VideoCapture(0)

# Setting propid's for width and height
cap.set(3,640)
cap.set(4,480)
while(True):
    ret, frame = cap.read()
    for barcode in decode(frame):
        print(barcode.data)
        value = barcode.data.decode("utf")
        print(value)
    cv2.imshow('result',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

# Setting propid's for width and height
cap.set(3,640)
cap.set(4,480)
while(True):
    ret, frame = cap.read()
    for barcode in decode(frame):
        value = barcode.data.decode("utf")
        print(value)
        pts = np.array([barcode.polygon])
        pts.reshape(-1,1,2)
        cv2.polylines(frame,[pts],True,(255,0,255),5)
        pts2 = barcode.rect
        cv2.putText(frame,value,(pts2[0],pts2[1]),cv2.FONT_HERSHEY_SIMPLEX,0.9,(255,0,255),2)
    cv2.imshow('result',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

### Authorization Based On QR Code

```python
cap = cv2.VideoCapture(0)
cap.set(3,640)
cap.set(4,480)
 
Allowed_persons = ["1111","1112","1113","1114","1115","1116","1117"]
 
while True:
 
    ret, frame = cap.read()
    for barcode in decode(frame):
        val = barcode.data.decode('utf-8')
        print(val)
 
        if val in Allowed_persons:
            myOutput = 'Authorized!'
            myColor = (0,255,0)
        else:
            myOutput = 'Un-Authorized Entry'
            myColor = (0, 0, 255)
 
        pts = np.array([barcode.polygon],np.int32)
        pts = pts.reshape((-1,1,2))
        cv2.polylines(frame,[pts],True,myColor,5)
        pts2 = barcode.rect
        cv2.putText(frame,myOutput,(pts2[0],pts2[1]),cv2.FONT_HERSHEY_SIMPLEX,0.9,myColor,2)
    cv2.imshow('result',frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    
    
cap.release()
cv2.destroyAllWindows() 
```
